```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load data
df = pd.read_csv("/content/Procurement KPI Analysis Dataset.csv")
print(" Data Loaded Successfully")
print(f"Shape: {df.shape}\n")

# --- Basic Info ---
print("--- Data Info ---")
print(df.info())
print("\n--- Missing Values ---")
print(df.isnull().sum())
print("\nMissing Value Percentage:\n", (df.isnull().sum()/len(df))*100)

# --- Target Variable ---
target_column = 'Compliance'   # Target for classification

# --- Handle Missing Values ---
# Fill missing Delivery_Date with most frequent date or Order_Date + mean delay
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], errors='coerce')
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

mean_delay = (df['Delivery_Date'] - df['Order_Date']).dt.days.mean()
df['Delivery_Date'] = df['Delivery_Date'].fillna(df['Order_Date'] + pd.to_timedelta(mean_delay, unit='d'))

# Fill Defective_Units with median
df['Defective_Units'] = df['Defective_Units'].fillna(df['Defective_Units'].median())

print("\nMissing values after imputation:\n", df.isnull().sum())

# --- Feature Engineering from Dates ---
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month
df['Order_Day'] = df['Order_Date'].dt.day
df['Delivery_Year'] = df['Delivery_Date'].dt.year
df['Delivery_Month'] = df['Delivery_Date'].dt.month
df['Delivery_Day'] = df['Delivery_Date'].dt.day
df['Delivery_Delay'] = (df['Delivery_Date'] - df['Order_Date']).dt.days

# Drop original date columns (keep derived numeric features)
df = df.drop(columns=['Order_Date', 'Delivery_Date', 'PO_ID'])

# --- Encode Categorical Variables ---
cat_cols = df.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
print("\nEncoded all categorical columns successfully.")

# --- Separate Features & Target ---
X = df.drop(columns=[target_column])
y = df[target_column]
y = LabelEncoder().fit_transform(y)  # Yes/No → 1/0

print("\nClass Distribution Before SMOTE:", Counter(y))

# --- Feature Scaling ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Apply SMOTE ---
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_scaled, y)

print("Class Distribution After SMOTE:", Counter(y_res))
print("\n✅ Data Preprocessing + SMOTE Balancing Completed Successfully!")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# -------------------------------
# Convert SMOTE output back to DataFrame for feature engineering
# -------------------------------

# Get column names from original X
feature_cols = X.columns
X_res_df = pd.DataFrame(X_res, columns=feature_cols)
y_res_df = pd.Series(y_res, name='Compliance')

# Combine features and target
df_smote = pd.concat([X_res_df, y_res_df], axis=1)
print("Shape after SMOTE (DataFrame):", df_smote.shape)

# -------------------------------
# Feature Engineering for Cold-Start Scenario
# -------------------------------

# 1. Aggregate / global features
df_smote['Avg_Order_Quantity'] = df_smote['Quantity'].mean()
df_smote['Avg_Defective_Units'] = df_smote['Defective_Units'].mean()
df_smote['Median_Delivery_Delay'] = df_smote['Delivery_Delay'].median()

# 2. Derived features
df_smote['Defective_Ratio'] = df_smote['Defective_Units'] / (df_smote['Quantity'] + 1e-6)
df_smote['Price_Difference'] = df_smote['Unit_Price'] - df_smote['Negotiated_Price']
df_smote['Delay_per_Quantity'] = df_smote['Delivery_Delay'] / (df_smote['Quantity'] + 1e-6)

# 3. One-hot encode categorical features for cold-start
# Identify categorical columns (already label-encoded, but for cold-start use one-hot)
cold_start_cat_cols = ['Item_Category', 'Order_Status']  # replace with actual categorical features
df_smote = pd.get_dummies(df_smote, columns=cold_start_cat_cols, drop_first=True)

# 4. Drop columns unavailable for cold-start
cols_to_drop = ['Supplier', 'PO_ID']  # supplier-specific columns
df_smote = df_smote.drop(columns=[col for col in cols_to_drop if col in df_smote.columns], errors='ignore')

# 5. Scale numeric features
numeric_cols = df_smote.select_dtypes(include=[np.number]).columns.drop('Compliance')
scaler = StandardScaler()
df_smote[numeric_cols] = scaler.fit_transform(df_smote[numeric_cols])

# -------------------------------
# Final Features & Target
# -------------------------------
X_cold = df_smote.drop(columns=['Compliance'])
y_cold = df_smote['Compliance']

print("\n✅ Cold-Start Feature Engineering on SMOTE Data Completed!")
print(f"Features shape: {X_cold.shape}, Target shape: {y_cold.shape}")
print("Sample features:\n", X_cold.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np

# -------------------------------
# Train-test split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_cold, y_cold, test_size=0.2, random_state=42, stratify=y_cold
)

# -------------------------------
# Define models
# -------------------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5)
}

# -------------------------------
# Train, Predict, and Evaluate
# -------------------------------
results = []

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)
    # Predict
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1] if hasattr(model, "predict_proba") else None

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    cr = classification_report(y_test, y_pred, digits=4)
    roc = roc_auc_score(y_test, y_prob) if y_prob is not None else "NA"

    results.append({
        "Model": name,
        "Accuracy": acc,
        "ROC-AUC": roc,
        "Confusion_Matrix": cm,
        "Classification_Report": cr
    })

# -------------------------------
# Display results
# -------------------------------
for res in results:
    print(f"\n=== {res['Model']} ===")
    print(f"Accuracy: {res['Accuracy']:.4f}")
    print(f"ROC-AUC: {res['ROC-AUC']}")
    print("Confusion Matrix:")
    print(res['Confusion_Matrix'])
    print("Classification Report:")
    print(res['Classification_Report'])
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# -------------------------------
# Define XGBoost model and hyperparameter grid
# -------------------------------
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=3,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# -------------------------------
# Fit GridSearch on training data
# -------------------------------
grid_search.fit(X_train, y_train)
print("Best XGBoost Parameters:", grid_search.best_params_)

# Use the best estimator
best_xgb = grid_search.best_estimator_

# -------------------------------
# Add XGBoost to models dictionary
# -------------------------------
models["XGBoost"] = best_xgb

# -------------------------------
# Train, Predict, Evaluate All Models Again (including XGBoost)
# -------------------------------
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1] if hasattr(model, "predict_proba") else None

    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    cr = classification_report(y_test, y_pred, digits=4)
    roc = roc_auc_score(y_test, y_prob) if y_prob is not None else "NA"

    results.append({
        "Model": name,
        "Accuracy": acc,
        "ROC-AUC": roc,
        "Confusion_Matrix": cm,
        "Classification_Report": cr
    })

# -------------------------------
# Display results
# -------------------------------
for res in results:
    print(f"\n=== {res['Model']} ===")
    print(f"Accuracy: {res['Accuracy']:.4f}")
    print(f"ROC-AUC: {res['ROC-AUC']}")
    print("Confusion Matrix:")
    print(res['Confusion_Matrix'])
    print("Classification Report:")
    print(res['Classification_Report'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
import pandas as pd

# -------------------------------
# Split original preprocessed data (before SMOTE) into train and test
# -------------------------------
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(
    X_cold, y_cold, test_size=0.2, random_state=42, stratify=y_cold
)

# -------------------------------
# Apply SMOTE only on training data
# -------------------------------
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_orig, y_train_orig)

print(f"Training data after SMOTE: {X_train_res.shape}, {y_train_res.shape}")
print(f"Testing data (original, imbalanced): {X_test_orig.shape}, {y_test_orig.shape}")

# -------------------------------
# Define models
# -------------------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": XGBClassifier(
        n_estimators=200, max_depth=7, learning_rate=0.1,
        subsample=0.8, colsample_bytree=0.8, eval_metric='logloss', random_state=42
    )
}

# -------------------------------
# Train, predict, evaluate on original test set
# -------------------------------
results = []

for name, model in models.items():
    # Train on SMOTE-resampled training data
    model.fit(X_train_res, y_train_res)

    # Predict on original test set (imbalanced)
    y_pred = model.predict(X_test_orig)
    y_prob = model.predict_proba(X_test_orig)[:,1] if hasattr(model, "predict_proba") else None

    acc = accuracy_score(y_test_orig, y_pred)
    cm = confusion_matrix(y_test_orig, y_pred)
    cr = classification_report(y_test_orig, y_pred, digits=4)
    roc = roc_auc_score(y_test_orig, y_prob) if y_prob is not None else "NA"

    results.append({
        "Model": name,
        "Accuracy": acc,
        "ROC-AUC": roc,
        "Confusion_Matrix": cm,
        "Classification_Report": cr
    })

# -------------------------------
# Display results
# -------------------------------
for res in results:
    print(f"\n=== {res['Model']} ===")
    print(f"Accuracy: {res['Accuracy']:.4f}")
    print(f"ROC-AUC: {res['ROC-AUC']}")
    print("Confusion Matrix:")
    print(res['Confusion_Matrix'])
    print("Classification Report:")
    print(res['Classification_Report'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -------------------------------
# Prepare data for visualization
# -------------------------------
model_names = [res['Model'] for res in results]
accuracy = [res['Accuracy'] for res in results]
roc_auc = [res['ROC-AUC'] if isinstance(res['ROC-AUC'], float) else 0 for res in results]
conf_matrices = [res['Confusion_Matrix'] for res in results]

# -------------------------------
# Bar chart: Accuracy and ROC-AUC
# -------------------------------
x = np.arange(len(model_names))
width = 0.35

fig, ax = plt.subplots(figsize=(12,6))
rects1 = ax.bar(x - width/2, accuracy, width, label='Accuracy')
rects2 = ax.bar(x + width/2, roc_auc, width, label='ROC-AUC')

ax.set_ylabel('Scores')
ax.set_title('Model Comparison: Accuracy vs ROC-AUC')
ax.set_xticks(x)
ax.set_xticklabels(model_names, rotation=45)
ax.set_ylim(0, 1.05)
ax.legend()

# Add value labels on top of bars
for i in range(len(x)):
    ax.text(i - width/2, accuracy[i]+0.02, f"{accuracy[i]:.2f}", ha='center')
    ax.text(i + width/2, roc_auc[i]+0.02, f"{roc_auc[i]:.2f}", ha='center')

plt.tight_layout()
plt.show()

# -------------------------------
# Side-by-side Confusion Matrices
# -------------------------------
fig, axes = plt.subplots(2, 3, figsize=(18,10))

for i, ax in enumerate(axes.flat):
    sns.heatmap(conf_matrices[i], annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(model_names[i])
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.suptitle('Confusion Matrices of All Models', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
